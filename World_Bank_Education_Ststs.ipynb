Importing all the libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading data and importing it into data frame
data = pd.read_csv("/content/drive/MyDrive/Data/data/EdStatsData.csv")
df = pd.DataFrame(data)
print(df.shape)


df = df.drop("Indicator Code", axis = 1)
```

```python
df["sum"] = df.sum(axis = 1)
df = df[df["sum"]!= 0]
df.shape
```

Total data was 8.8L and the data which has no null values are just 3.5L
so these will be data which we will analyse

```python
list(df["Indicator Name"].unique())
```

```python
primary_df = df[df["Indicator Name"] == 'Wittgenstein Projection: Mean years of schooling. Age 25+. Total']
primary_df = primary_df.reset_index(drop = True)
```

```python
primary_df = primary_df.sort_values(by = '2010', ascending=True)[80:]
```

```python
fig, plot = plt.subplots(figsize=(20,30))
mygraph = plot.barh(primary_df['Country Name'], primary_df['2010'])
plt.axvline(x = np.mean(primary_df['2010']), color = 'green', linewidth = 1.5,)
plt.title('Average years of Total Schooling in 2010',fontsize=30,fontweight='bold')
plt.xlabel("Number of years of schooling")
plt.ylabel("Countries")
plt.show()
```
